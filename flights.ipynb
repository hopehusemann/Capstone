```python
import pandas as pd

# Load dataset
flight_data = pd.read_csv(r"C:\Users\hopeh\Desktop\data_science_bootcamp\flight_times_capstone\flight_data.csv", low_memory=False)

# Display the first few rows of the dataset
print(flight_data.head())

# Display basic information about the dataset
print(flight_data.info())

# Show summary statistics
print(flight_data.describe(include='all'))
```

```python
# Display column names
print(flight_data.columns)
```

```python
# Reformat column titles

# Convert all column names to lowercase
flight_data.columns = flight_data.columns.str.lower()

# Rename specific columns if needed
flight_data.rename(columns={
    'fl_date': 'date', 
    'origin_city_name': 'origin_city',
    'dest_city_name' : 'dest_city',
    'origin' : 'origin_iata',
    'dest' : 'dest_iata',
}, inplace=True)
```

```python
# Display column names
print(flight_data.columns)
```

```python
# Convert float64 to float32
for col in flight_data.select_dtypes(include=['float64']).columns:
    flight_data[col] = flight_data[col].astype('float32')

# Check memory usage after conversion
print(flight_data.info())
```

```python
# Calculate the proportion of missing values for each column
missing_proportion = flight_data.isnull().mean()

# Filter columns where the proportion of missing values is more than 20%
high_missing_cols = missing_proportion[missing_proportion > 0.20].index

# Display columns with high missing values
print("Columns with more than 20% missing values:")
print(high_missing_cols)
```

```python
# Calculate the percentage of NaN values for each column
nan_percentage = flight_data.isna().mean() * 100

# Adjust pandas display options
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows if there are many rows

# Print the results
print(nan_percentage)
```

```python
# Drop columns with more than 95% missing values

# Identify columns with more than 95% missing values
columns_to_drop = nan_percentage[nan_percentage > 95.0].index

# Drop the identified columns
flight_data = flight_data.drop(columns=columns_to_drop)

# Verify the changes
print(flight_data.info())
```

```python
# Calculate the percentage of NaN values for each column
nan_percentage = flight_data.isna().mean() * 100

# Adjust pandas display options
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows if there are many rows

# Print the results
print(nan_percentage)
```

All of the columns with nan percentages of approximately 79.32 appear to be related to delays specifically so we are going to investigate what this data is before considering removal.

```python
# Calculate the number of NaN values per row
num_nans_per_row = flight_data.isnull().sum(axis=1)

# Calculate the total number of columns in the DataFrame
total_columns = flight_data.shape[1]

# Calculate the proportion of NaN values per row
nan_proportion_per_row = num_nans_per_row / total_columns
```

```python
# Define the threshold (e.g., 80%)
threshold = 0.80

# Identify rows where the proportion of NaN values is more than the threshold
rows_with_high_nan = flight_data[nan_proportion_per_row > threshold]

# Display the rows with high NaN proportions
print("Rows with more than 80% NaN values:")
print(rows_with_high_nan)
```

```python
# Drop Columns that provide no useful information for analysis (e.g., unique
# identifiers if not needed)

columns_to_drop = [
    'origin_city_market_id',
    'dest_city_market_id',
    'origin_state_nm',
    'dest_state_nm', 
    'origin_airport_id',
    'dest_airport_id'
]

# Drop the identified columns
flight_data = flight_data.drop(columns=columns_to_drop)

# Verify the changes
print(flight_data.info())
```

```python
print(flight_data.head())
```

```python
# Split origin_city column to reflect city and state
flight_data[['origin_city', 'origin_state']] = flight_data['origin_city'].str.split(', ', expand=True)

# Split dest_city column to reflect city and state
flight_data[['dest_city', 'dest_state']] = flight_data['dest_city'].str.split(', ', expand=True)
```

```python
print(flight_data.head())
```

```python
# Drop rows with missing values in year, month, or day_of_month
flight_data = flight_data.dropna(subset=['year', 'month', 'day_of_month'])
```

```python
# check data types
flight_data.info()
```

```python
# Check for invalid day values
invalid_days = flight_data[
    (flight_data['day_of_month'] < 1) | 
    (flight_data['day_of_month'] > 31)
]
print("\nInvalid day_of_month values:")
print(invalid_days)
```

```python
# Check for duplicate rows
# duplicates = flight_data[flight_data.duplicated()]
# print("\nDuplicate rows:")
# print(duplicates)

duplicate_count = flight_data.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")
```

```python
# Identify duplicate rows
duplicates = flight_data[flight_data.duplicated(keep=False)]

# Display the first 10 duplicate rows
print(duplicates.head(10))
```

```python
# Normalize string columns (remove spaces and convert to lower case)
for col in flight_data.select_dtypes(include='object').columns:
    flight_data[col] = flight_data[col].str.strip().str.lower()

# Identify duplicate rows where all cells are identical
duplicates = flight_data[flight_data.duplicated(keep=False)]
```

```python
# Display the first 10 duplicate rows
print(duplicates.head(10))
```

```python
# Count the number of total duplicates
duplicate_count = duplicates.shape[0]
print(f"Total duplicate rows (all cells identical): {duplicate_count}")
```

```python
# Drop all duplicate rows
flight_data = flight_data.drop_duplicates()
```

```python
# Check duplicates
duplicate_count = flight_data.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")
```

```python
# combine the three columns into a new 'date' column in datetime format
# Ensure to handle any missing values as needed
#flight_data['date'] = pd.to_datetime(flight_data[['year', 'month', 'day_of_month']], errors='coerce')

# Creating the date column using apply
flight_data['date'] = flight_data.apply(lambda row: pd.Timestamp(year=row['year'], month=row['month'], day=row['day_of_month']), axis=1)

# Display the updated DataFrame
print("\nUpdated DataFrame with date:")
print(flight_data.head())
```
